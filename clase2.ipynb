```python
def testPrime(X):          #Si X es primo, testPrime(X)=True, si X no es primo, testPrime(X)=False
    
    if type(X)!=int:       #esto es por si X no es un int
        return(False)
    
    elif X<2:              #con esto no tengo que lidear con números menores a 2
        return(False)
    else:                  
        Z=True             #Z es el resultado, empieza siendo True
        for n in range(X-2): # en este loop veo si Z es divisible en todos los números desde el 2 hasta X-1
            if X%(n+2)==0:   # si X si es divisible, entonces Z cambia a False y el loop se detiene
                Z=False
                break
        return(Z)

def prime(X):               #esta función da una lista con los números primos desde 2 hasta X 
    Z=[]                     #Z es la lista que da la función al final, empieza estando vacía
    if type(X)==int:           #esto es para que X sea un entero, si no lo es, regresa una lista vacía
        for a in range(X):      #En esta ciclo se prueban todos los números desde 1 hasta X
            if testPrime(a+1):   #si son primos se agregan a Z
                Z.append(a+1)
    return(Z)

def twinPrime(X):          #esta función regresa un lista con pares de primos gemelos, 
    Z=[]                
    if type(X)==int:        #primero verifica que x sea un entero
        for a in range(X):  
            if testPrime(a+1) and testPrime(a-1):  #cuando encuentra que un número (a+1) es primo checa que (a+1)-2
                Z.append([a-1,a+1])                  #sea primo también. si esto ocurre los agrega a Z en una lista
        return(Z)        
    
def theoremArihtmetic(X):  #esta función regresa una lista Z=[ZP,ZN] donde: 
    if type(X)==int:
        ZP=prime(X)            #ZP es una lista que contiene numeros primos desde 2 hasta X
        ZN=[]                  #ZN es una lista contiene a las potencias correspondientes a cada primo
        for a in range(len(ZP)):#en este ciclo se llena a ZN de 0s para que len(ZN)=len(ZP)
            ZN.append(0)
        while X>1:               #este ciclo checa si X es divisible entre los elementos de ZP
            for a in range(len(ZP)):    
                if X%ZP[a]==0:   #si es divisible le suma 1 al elemento de ZN correspondiente y hace que X=X/elemento de (ZP)
                    ZN[a]=ZN[a]+1 #el ciclo while termina cuando X==1
                    X=X/ZP[a]              
                    break   #este break es para que no haga cuentas innecesarias
        return([ZP,ZN])

def mcm(X,Y): #esta función regresa al minimo común múltiplo de X y Y
    PX=theoremArihtmetic(X)[0] #PX son los números primos desde 2 hasta X
    PY=theoremArihtmetic(Y)[0] #PY son los números primos desde 2 hasta Y
    NX=theoremArihtmetic(X)[1] #NX son las potencias asociadas a los primos que forman a X
    NY=theoremArihtmetic(Y)[1] #NY son las potencias asociadas a los primos que forman a Y
    
    if len(PX)>len(PY):           #en este if y else hago que PZ sea igual a la lista mas larga de primos
        PZ=PX                     #PZ son los primos que componen al mcm llamado Z
        while len(NX)>len(NY):    #también se le agregan 0s al final de la menor de las listas NX y NY para que tengan 
            NY.append(0)          #el mismo número de elementos
    else:
        PZ=PY
        while len(NY)>len(NX):
            NX.append(0)
    NZ=[]                        #NZ es la lista que contiene a las potencias de los primos que forman a Z
    for a in range(len(NX)):     #cada elemento de NZ es el mayor elemento entre NX y NY
        if NX[a]>NY[a]:         
            NZ.append(NX[a])
        else:                    
            NZ.append(NY[a])
    Z=1                        #Z es el mcm de X y Y, y es igal al producto de las potencias de los primos obtenidos
    for a in range(len(NZ)):   
        Z=Z*((PZ[a])**(NZ[a]))
    
    return(Z)
def MCD(X,Y):                  #esta funcion es igual a Z, el maximo comun divisor de X y Y
    PX=theoremArihtmetic(X)[0] #PX son los números primos desde 2 hasta X
    PY=theoremArihtmetic(Y)[0] #PY son los números primos desde 2 hasta Y
    NX=theoremArihtmetic(X)[1] #NX son las potencias asociadas a los primos que forman a X
    NY=theoremArihtmetic(Y)[1] #NY son las potencias asociadas a los primos que forman a Y
    
    if len(PX)<len(PY):        #PZ es la lista que contiene a los primos que forman a Z
        PZ=PX                  #en este if y else hago que PZ sea igual a la lista mas corta de primos
        while len(NX)<len(NY): #y para que NX y NY tengan el mismo número de elementos le quito los últimos a la lista más larga
            NY.pop(len(NY)-1) 
    else:
        PZ=PY
        while len(NY)<len(NX):
            NX.pop(len(NX)-1)
    NZ=[]                     #NZ tiene a las potnecias de los primos que forman a Z
    for a in range(len(NX)):  #En este ciclo for hago que los elementos de NZ sean los más pequeños de NX y NY
        if NX[a]<NY[a]:
            NZ.append(NX[a])
        else:
            NZ.append(NY[a])
    Z=1
    for a in range(len(NZ)):    #Z es el mcm de X y Y, y es igal al producto de las potencias de los primos obtenidos
        Z=Z*((PZ[a])**(NZ[a]))
    
    return(Z)

```
